```python
import pandas as pd

tweet = pd.read_csv('../kaggle_data/merged_tweets.csv',
                       usecols=['ticker_symbol', 'Date', 'body', 'comment_num', 'retweet_num', 'like_num'])
tweet = tweet[(tweet["Date"]>='2019-01-01') & (tweet["Date"]<'2019-12-30')]
```

```python
tweets = tweet
tweets["total_engagement"] = tweets["comment_num"] + tweets["retweet_num"] + tweets["like_num"]
```

```python
tweets.sort_values(by='Date', ascending=True, inplace=True)
tweets
```

```python
tweets.to_csv("page1-1.csv", index=False)
```

```python
from pysentimiento.preprocessing import preprocess_tweet
from pysentimiento import SentimentAnalyzer

analyzer = SentimentAnalyzer(lang="en")
def cleanUpTweet(txt):
    txt = preprocess_tweet(txt, lang="en", shorten=2)
    return txt

def getOutput(txt):
    return analyzer.predict(txt)

def getProb(output):
    return output.probas[output.output]

def getTextClass(output):
    return output.output
```

```python
tweets["clean_text"] = tweets["body"].apply(cleanUpTweet)
```

```python
tweets["prob"] = tweets["clean_text"].apply(getOutput).apply(getProb)
```

```python
tweets["tweet_class"] = tweets["clean_text"].apply(getOutput).apply(getTextClass)
```
